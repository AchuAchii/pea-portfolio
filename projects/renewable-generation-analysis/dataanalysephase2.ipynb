```python
import pandas as pd
from google.colab import files
uploaded = files.upload()
```

```python
import numpy as np
file_name = list(uploaded.keys())[0]
df = pd.read_csv(file_name)
df = df[df["เชื้อเพลิง"].str.contains("ชีวภาพ", case=False, na=False)]
df.head()
```

```python
df['AVG_EFF'] = df.groupby('id')['efficiency'].transform('mean')
df['AVG'] = df.groupby('id')['AVG_kWH'].transform('mean')
df['STD'] = df.groupby('id')['AVG_kWH'].transform('std')
df['CV'] = (df['STD'] / df['AVG']) * 100
aggregated_df = df.groupby(['id'], as_index=False).apply(lambda g: g.iloc[0]).reset_index(drop=True)
aggregated_df = aggregated_df.fillna(0)
aggregated_df['AVG_EFF'] = aggregated_df['AVG_EFF'].round(2)
aggregated_df['CV'] = aggregated_df['CV'].round(2)

aggregated_df.head()
```

```python
cols = [col for col in aggregated_df.columns if col != 'missing_rate_kWH'] + ['missing_rate_kWH']
aggregated_df = aggregated_df[cols]
```

```python
aggregated_df.drop(columns=['efficiency','month','mwตามสัญญา','mwติดตั้ง','AVG_kWH','PF100','AVG','STD']).to_csv('1.csv', index=False)
```
