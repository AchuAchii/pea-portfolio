```python
import pandas as pd
import numpy as np  

df = pd.read_csv("2567.csv", encoding='utf-8')  

df = df.sort_values(by='ID')
df.head() 
```

```python
days_in_month = {
    'JAN': 31, 'FEB': 28, 'MAR': 31, 'APR': 30, 
    'MAY': 31, 'JUN': 30, 'JUL': 31, 'AUG': 31,
    'SEP': 30, 'OCT': 31, 'NOV': 30, 'DEC': 31
}
for month, days in days_in_month.items():
    df[f'PF_{month}'] = days * df['mw ตามสัญญา'] * 24000

df.head()
```

```python
df['AVG_PF'] = df[[f'PF_{month}' for month in days_in_month.keys()]].mean(axis=1)
df.head()
```

```python
print("Column names:", df.columns.tolist())
```

```python
df['PFC_JAN67'] = (df[' Jan ']/df['PF_JAN'] ) * 100
df['PFC_FEB67'] = (df[' Feb ']/df['PF_FEB'] ) * 100
df['PFC_MAR67'] = (df[' Mar ']/df['PF_MAR'] ) * 100
df['PFC_APR67'] = (df[' Apr ']/df['PF_APR'] ) * 100
df['PFC_MAY67'] = (df[' May ']/df['PF_MAY'] ) * 100
df['PFC_JUN67'] = (df[' Jun ']/df['PF_JUN'] ) * 100
df['PFC_JUL67'] = (df[' Jul ']/df['PF_JUL'] ) * 100
df['PFC_AUG67'] = (df[' Aug ']/df['PF_AUG'] ) * 100
df['PFC_SEP67'] = (df[' Sep ']/df['PF_SEP'] ) * 100
df['PFC_OCT67'] = (df[' Oct ']/df['PF_OCT'] ) * 100
df['PFC_NOV67'] = (df[' Nov ']/df['PF_NOV'] ) * 100
df['PFC_DEC67'] = (df[' Dec ']/df['PF_DEC'] ) * 100

df.head()
```

```python
df.drop(columns=['PF_JAN', 'PF_FEB', 'PF_MAR', 'PF_APR', 'PF_MAY', 'PF_JUN', 'PF_JUL', 'PF_AUG',
                'PF_SEP', 'PF_OCT', 'PF_NOV', 'PF_DEC','AVG_PF']).to_csv("2567test.csv", index=False, encoding='utf-8-sig')
```
