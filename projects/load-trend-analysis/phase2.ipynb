```python
import pandas as pd
from functools import reduce

files = [
    ("2560test.csv", "y2560"),
    ("2561test.csv", "y2561"),
    ("2562test.csv", "y2562"),
    ("2563test.csv", "y2563"),
    ("2564test.csv", "y2564"),
    ("2565test.csv", "y2565"),
    ("2566test.csv", "y2566"),
    ("2567test.csv", "y2567"),
]

merge_keys = ["ID", "เขต", "mw ตามสัญญา", "mw ติดตั้ง", "วันที่ cod"]

dfs = []
for f, prefix in files:
    df = pd.read_csv(f, encoding='utf-8')
    new_cols = {
        c: f"{prefix}_{c}" for c in df.columns if c not in merge_keys
    }
    df = df.rename(columns=new_cols)
    dfs.append(df)

merged_df = reduce(lambda left, right: pd.merge(left, right, on=merge_keys, how="outer"), dfs)

merged_df.to_csv("test01.csv", index=False, encoding='utf-8-sig')
print("✅ รวมไฟล์สำเร็จแล้ว -> test01.csv")
```
