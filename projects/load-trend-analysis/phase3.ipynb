```python
# 1. แสดงชื่อคอลัมน์ทั้งหมด
print("ชื่อคอลัมน์ทั้งหมด:", df.columns.tolist())

# 2. สร้างฟังก์ชันค้นหาชื่อคอลัมน์
def find_month_columns(year, month, df_columns):
    patterns = [
        f'y{year}_{month}{str(year)[2:]}',
        f'y{year}_ {month}{str(year)[2:]}',
        f'y{year} {month}{str(year)[2:]}',
        f'y{year}{month}{str(year)[2:]}'
    ]
    for pattern in patterns:
        matches = [col for col in df_columns if col.replace(' ', '') == pattern.replace(' ', '')]
        if matches:
            return matches[0]
    return None

# 3. รวบรวมคอลัมน์ที่ถูกต้อง
valid_columns = []
missing_columns = []

for year in range(2560, 2568):
    for month in ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
                  'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']:
        col = find_month_columns(year, month, df.columns)
        if col:
            valid_columns.append(col)
        else:
            missing_columns.append(f"y{year}_{month}{str(year)[2:]}")

# 4. คำนวณ missing rate (รวมค่า 0.0)
if valid_columns:
    # นับทั้ง NaN และค่า 0.0
    missing_count = ((df[valid_columns].isna()) | (df[valid_columns] == 0.0)).sum(axis=1)
    df['missing_rates_All'] = (missing_count)
    
    # แสดงผลลัพธ์
    print(f"\nจำนวนคอลัมน์ทั้งหมด: {len(valid_columns)} คอลัมน์")
    print("\nตัวอย่างผลการคำนวณ missing rate (รวมค่า 0.0):")
    print(df[['ID', 'missing_rates_All']].head())
    
    # แสดงสถิติเพิ่มเติม
    print("\nสถิติของ missing rate:")
    print(f"ค่าเฉลี่ย: {df['missing_rates_All'].mean():.2f}%")
    print(f"ค่าต่ำสุด: {df['missing_rates_All'].min():.2f}%")
    print(f"ค่าสูงสุด: {df['missing_rates_All'].max():.2f}%")
else:
    print("ไม่พบคอลัมน์ที่ต้องการ")
```
